```python
from urllib.request import urlopen as ureq
from bs4 import BeautifulSoup as soup
from requests_html import HTMLSession
```

```python
my_url = 'https://paytmmall.com/allsmartphones-llpid-195328?use_mw=1&src=store&from=storefront&tracker=%7C%7C%7C%7C%2Fh%2Fmobiles-tablets-clpid-6942-Shop%20By%20Category%7C310885%7C1%7C%7C%7C%7C'
```

```python

uclient= ureq(my_url)
page_html= uclient.read()
uclient.close()
```

```python
page_soup= soup(page_html, 'html.parser')
page_soup
```

```python
containers= page_soup.find_all('div', {"class":"_2i1r"})
print(len(containers))
```

```python
containers
```

```python
filename= "payTm_mall.csv"
f=open(filename, "w")
headers= 'Name, price, Discount \n'
f.write(headers)

for items in containers:
    Name = items.find_all("div", {"class": "UGUy"})
    Name=Name[0].text
    price= items.find("div", {"class": "_1kMS"})
    price=price.text.replace(',', '')
    Discount_tag= items.findAll("span", {"class": "c-ax"})
    try:
        Discount = Discount_tag[0].text
    except:
        Discount="NA"
    print(Name + "," + price + "," + Discount + "\n")
    f.write(Name + "," + price + "," + Discount + "\n")


    
```

```python
f.close()
```
